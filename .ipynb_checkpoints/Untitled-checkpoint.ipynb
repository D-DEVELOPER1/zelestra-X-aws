```python
test_data = pd.read_csv("dataset/test.csv")
train_data = pd.read_csv("dataset/train.csv")
merge_data = pd.merge(test_data, train_data, on=["id"])
merge_data.head()
```

```python
for label, content in merge_data.items():
    if pd.api.types.is_string_dtype(content):
        merge_data[label] = content.astype("category").cat.as_ordered()
```

```python
for label, content in merge_data.items():
    if pd.api.types.is_numeric_dtype(content):
        print(label)
```

```python
for label, content in merge_data.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            merge_data[label] = content.fillna(content.median())
```

```python
for label, content in merge_data.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            print(label)
```

```python
for label, content in merge_data.items():
    if not pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            print(label)
```

```python
for label, content in merge_data.items():
    if not pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            merge_data[label] = pd.Categorical(content).codes + 1
```

```python
merge_data.isna().sum()
```

```python
x = merge_data.drop("efficiency", axis = 1)
y = merge_data["efficiency"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
x_test, x_train, y_test, y_train = train_test_split(x,
                                                    y,
                                                    test_size = 0.2)
```

```python
model = RandomForestRegressor(n_estimators = 100)
model.fit(merge_data.drop("efficiency", axis = 1), merge_data["efficiency"])
model.score(merge_data.drop("efficiency", axis = 1), merge_data["efficiency"])
```
